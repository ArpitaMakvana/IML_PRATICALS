```python

import matplotlib.pyplot as plt
stock_prices = pd.DataFrame({'open': [60, 70, 80, 90, 100, 110, 120],
   'close': [55, 85, 75, 100, 95, 120, 105],
   'high': [70, 95, 85, 110, 105, 120, 125],
   'low': [50, 60, 70, 80, 85, 105, 100]},
                index=pd.date_range("2023-04-1",periods=7,freq="d"))

plt.figure()
up = stock_prices[stock_prices.close >= stock_prices.open]
down = stock_prices[stock_prices.close < stock_prices.open]

col1="black"
col2="yellow"

width = 0.4
width2 =0.05

plt.bar(up.index, up.close-up.open, width, bottom=up.open, color=col2)
plt.bar(up.index, up.high-up.close, width2, bottom=up.close, color=col2)
plt.bar(up.index, up.low-up.open, width2, bottom=up.open, color=col2)

plt.bar(down.index, down.close-down.open, width, bottom=down.open, color=col1)
plt.bar(down.index, down.high-down.open, width2, bottom=down.open, color=col1)
plt.bar(down.index, down.low-down.close, width2, bottom=down.close, color=col1)

plt.xticks(rotation=45, ha='right')


plt.title('Stock Prices for a Week')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('/kaggle/input/nifty50-stock-market-data/ADANIPORTS.csv')


# Limit the DataFrame to the first 130 rows
limited_df = df.iloc[:12]

# Create a date range with the desired start date and frequency
date_range = pd.date_range("2000-03-07", periods=len(limited_df), freq="D")

# Set the date range as the index of the limited DataFrame
limited_df.set_index(date_range, inplace=True)

# Define colors for candlesticks
col1 = 'black'  
col2 = 'yellow'    

# Define width for candlestick elements
width = .3
width2 =.03

# Create separate DataFrames for "up" (increasing) and "down" (decreasing) pricesw
up = limited_df[limited_df['Close'] > limited_df['Open']]
down = limited_df[limited_df['Open'] > limited_df['Close']]

# Plotting up prices of the stock
plt.bar(up.index, up['Close'] - up['Open'], width, bottom=up['Open'], color=col1)
plt.bar(up.index, up['High'] - up['Close'], width2, bottom=up['Close'], color=col1)
plt.bar(up.index, up['Low'] - up['Open'], width2, bottom=up['Open'], color=col1)

# Plotting down prices of the stock
plt.bar(down.index, down['Open'] - down['Close'], width, bottom=down['Close'], color=col2)
plt.bar(down.index, down['Open'] - down['High'], width2, bottom=down['High'], color=col2)
plt.bar(down.index, down['Open'] - down['Low'], width2, bottom=down['Low'], color=col2)


plt.xticks(rotation=30, ha='right')


plt.show()
```
